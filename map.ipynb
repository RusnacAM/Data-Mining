```python
# Sources: https://datascience.quantecon.org/tools/maps.html, https://stackoverflow.com/questions/72249181/how-to-insert-points-in-a-map
import pandas as pd
import folium as folium
import random
```

```python
df = pd.read_csv("C:\\Users\\Adrian\\Documents\\Aarhus University\\Data Mining\\Gowalla Data\\Gowalla_totalCheckins.txt", sep = "\t", header = None)
# read csv is still using local file path.
headers = ["user_id", "checkin_time", "latitude", "longitude", "location_id"]
df.columns = headers
print(df.head())
```

```python
# Creating the map
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
```

```python
random.seed(1)
mylist = list(range(0, len(df)))
randomlist = random.sample(mylist, k=1000) # k can be set higher if you want more points
```

```python
print(df['location_id'].nunique())
```

```python
#for i in df.loc[(df['user_id'] == 145)].index:
for i in randomlist:
#for i in df['location_id'].unique():
    folium.Marker(
      location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
      popup=df.iloc[i]['user_id'],
   ).add_to(m)

m
```
