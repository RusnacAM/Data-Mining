```python
# Sources: https://datascience.quantecon.org/tools/maps.html, https://stackoverflow.com/questions/72249181/how-to-insert-points-in-a-map
import pandas as pd
import folium as folium
```

```python
df = pd.read_csv("C:\\Users\\Adrian\\Data-Mining\\data\\Gowalla_totalCheckins.txt", sep = "\t", header = None)
# read csv is still using local file path.
headers = ["user_id", "checkin_time", "latitude", "longitude", "location_id"]
df.columns = headers
print(df.head())
```

```python
# Creating the map
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
```

```python
for i in range(0,len(df)):
    folium.Marker(
      location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
      popup=df.iloc[i]['user_id'],
   ).add_to(m)

m
```
